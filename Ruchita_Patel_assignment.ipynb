```python
import json
import math
import random
import csv
```

```python
# Input JSON Data
with open("/content/drive/MyDrive/python_Ruchita/input/data.json", "r") as file:
    data = json.load(file)
    print(data)
```

```python
warehouses = data["warehouses"]
agents = data["agents"]
packages = data["packages"]
```

```python
def distance(p1, p2):
    return math.sqrt(
        (p1[0] - p2[0]) ** 2 +
        (p1[1] - p2[1]) ** 2
    )


#  New Agent Mid-Day
agents["A4"] = [20, 20]

print("New Agent Joined Mid-Day -> A4\n")

# ASCII Visualization

print("ASCII VISUALIZATION")
print("-" * 40)

for w, loc in warehouses.items():
    print(f"{w} Warehouse -> {loc}")

for a, loc in agents.items():
    print(f"{a} Agent      -> {loc}")

print("-" * 40)


# Create Report
report = {}

for agent in agents:
    report[agent] = {
        "packages_delivered": 0,
        "total_distance": 0
    }


# Process Deliveries

for package in packages:

    warehouse_id = package["warehouse"]
    destination = package["destination"]

    warehouse_location = warehouses[warehouse_id]

    nearest_agent = None
    minimum_distance = float("inf")

    # Find nearest agent
    for agent_id, agent_location in agents.items():

        d = distance(agent_location, warehouse_location)

        if d < minimum_distance:
            minimum_distance = d
            nearest_agent = agent_id

    # Distance warehouse destination
    delivery_distance = distance(
        warehouse_location,
        destination
    )

    # Random delivery delay
    random_delay = random.randint(1, 5)

    total_distance = (
        minimum_distance +
        delivery_distance +
        random_delay
    )

    # Update report
    report[nearest_agent]["packages_delivered"] += 1
    report[nearest_agent]["total_distance"] += total_distance

    # Print route
    print(f"\nPackage : {package['id']}")
    print(f"Agent   : {nearest_agent}")
    print(f"Route   : {nearest_agent} -> {warehouse_id} -> {destination}")
    print(f"Delay   : {random_delay} km")
    print(f"Distance: {round(total_distance, 2)}")

# Calculate Efficiency

best_agent = None
best_efficiency = float("inf")

for agent, details in report.items():

    delivered = details["packages_delivered"]

    if delivered > 0:
        efficiency = (
            details["total_distance"] / delivered
        )
    else:
        efficiency = 0

    details["total_distance"] = round(
        details["total_distance"], 2
    )

    details["efficiency"] = round(
        efficiency, 2
    )

    if delivered > 0 and efficiency < best_efficiency:
        best_efficiency = efficiency
        best_agent = agent

# Add best agent
report["best_agent"] = best_agent

print("\nFINAL REPORT")
print(json.dumps(report, indent=4))
```

```python
# Save JSON Report

with open("/content/drive/MyDrive/python_Ruchita/output/report.json", "w") as file:
    json.dump(report, file, indent=4)

print("\nreport.json saved")
```

```python
# Export Best Agent to CSV

with open("/content/drive/MyDrive/python_Ruchita/output/top_performer.csv", "w", newline="") as file:

    writer = csv.writer(file)

    writer.writerow([
        "Agent",
        "Packages Delivered",
        "Total Distance",
        "Efficiency"
    ])

    writer.writerow([
        best_agent,
        report[best_agent]["packages_delivered"],
        report[best_agent]["total_distance"],
        report[best_agent]["efficiency"]
    ])

print("top_performer.csv saved")
```
